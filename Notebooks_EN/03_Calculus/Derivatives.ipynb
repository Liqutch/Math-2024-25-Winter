1. Compute derivatives of functions:
   * $y(x) = -3x+3$
   * $y(x) = \pi x + \sin(1)$
   * $y(x) = 4+\sin(2)$
   * $y(x) = 2x^3 - 3x^2 + 8x - 9$
   * $y(x) = 6 x^{1/3}$
   * $y(x) = \sqrt{x}$
   * $y(x) = \cos(x) + \sin(x)$
   * $y(x) = 2\sin(x) \cos(x)$
   * $y(x) = x\sin(x)$
   * $y(x) = (x+1)(x+1)$
   * $y(x) = \frac{x}{x+1}$
   * $y(x) = (x+1)\exp(x)$
   * $y(x) = \sin(x^2)$
   * $y(x) = \exp(-2x)$
   * $y(x) = \frac{1}{\sin(x+1)}$
   * $y(x) = \sqrt{2x+1}$

### Answers of Exercise #1

1. **\( y(x) = -3x + 3 \):**
   $$
   y'(x) = -3
   $$

2. **\( y(x) = \pi x + \sin(1) \):**
   - \( \sin(1) \) is a constant, so its derivative is 0.
   $$
   y'(x) = \pi
   $$

3. **\( y(x) = 4 + \sin(2) \):**
   - Both terms are constants, so:
   $$
   y'(x) = 0
   $$

4. **\( y(x) = 2x^3 - 3x^2 + 8x - 9 \):**
   Using the power rule:
   $$
   y'(x) = 6x^2 - 6x + 8
   $$

5. **\( y(x) = 6x^{1/3} \):**
   Using the power rule:
   $$
   y'(x) = 6 \cdot \frac{1}{3} x^{-2/3} = 2x^{-2/3}
   $$

6. **\( y(x) = \sqrt{x} \):**
   Rewrite \( \sqrt{x} \) as \( x^{1/2} \), then use the power rule:
   $$
   y'(x) = \frac{1}{2}x^{-1/2} = \frac{1}{2\sqrt{x}}
   $$

7. **\( y(x) = \cos(x) + \sin(x) \):**
   Using the derivatives of trigonometric functions:
   $$
   y'(x) = -\sin(x) + \cos(x)
   $$

8. **\( y(x) = 2\sin(x)\cos(x) \):**
   Use the product rule:
   $$
   y'(x) = 2\cos^2(x) - 2\sin^2(x)
   $$
   Alternatively, simplify using \( \sin(2x) \):
   $$
   y'(x) = \sin(2x)
   $$

9. **\( y(x) = x\sin(x) \):**
   Use the product rule:
   $$
   y'(x) = 1 \cdot \sin(x) + x \cdot \cos(x) = \sin(x) + x\cos(x)
   $$

10. **\( y(x) = (x+1)(x+1) \):**
    Rewrite as \( y(x) = (x+1)^2 \) and use the chain rule:
    $$
    y'(x) = 2(x+1)
    $$

11. **\( y(x) = \frac{x}{x+1} \):**
    Use the quotient rule:
    $$
    y'(x) = \frac{(1)(x+1) - (x)(1)}{(x+1)^2} = \frac{1}{(x+1)^2}
    $$

12. **\( y(x) = (x+1)\exp(x) \):**
    Use the product rule:
    $$
    y'(x) = (1)(\exp(x)) + (x+1)(\exp(x)) = \exp(x) + (x+1)\exp(x) = (x+2)\exp(x)
    $$

13. **\( y(x) = \sin(x^2) \):**
    Use the chain rule:
    $$
    y'(x) = \cos(x^2) \cdot 2x = 2x\cos(x^2)
    $$

14. **\( y(x) = \exp(-2x) \):**
    Use the chain rule:
    $$
    y'(x) = \exp(-2x) \cdot (-2) = -2\exp(-2x)
    $$

15. **\( y(x) = \frac{1}{\sin(x+1)} \):**
    Rewrite as \( y(x) = (\sin(x+1))^{-1} \) and use the chain rule:
    $$
    y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}
    $$

16. **\( y(x) = \sqrt{2x+1} \):**
    Rewrite as \( y(x) = (2x+1)^{1/2} \) and use the chain rule:
    $$
    y'(x) = \frac{1}{2}(2x+1)^{-1/2} \cdot 2 = \frac{1}{\sqrt{2x+1}}
    $$

2. Prove:
   - $\frac{d}{dx} (\ln(\sin(x))) = \cot(x)$


Start with the derivative of the natural logarithm:
   $$
   \frac{d}{dx} (\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \frac{d}{dx}(\sin(x))
   $$

The derivative of \( \sin(x) \) is \( \cos(x) \):
   $$
   \frac{d}{dx} (\ln(\sin(x))) = \frac{\cos(x)}{\sin(x)}
   $$
Simplify \( \frac{\cos(x)}{\sin(x)} \) to \( \cot(x) \):
   $$
   \frac{d}{dx} (\ln(\sin(x))) = \cot(x)
   $$

3. For $f(x) = \cos(x)$, verify that $f''(x) = -f(x)$.

### Answers of Exercise #3

Compute the first derivative:
   $$
   f'(x) = \frac{d}{dx} (\cos(x)) = -\sin(x)
   $$
Compute the second derivative:
   $$
   f''(x) = \frac{d}{dx} (-\sin(x)) = -\cos(x)
   $$
Compare \( f''(x) \) with \( -f(x) \):
   Since \( f(x) = \cos(x) \), we find:
   $$
   f''(x) = -f(x)
   $$
